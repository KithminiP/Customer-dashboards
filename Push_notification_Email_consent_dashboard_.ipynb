```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample data - replace this with your actual data
data = {
    'week': ['2024-01', '2024-02', '2024-03', '2024-04', '2024-05'],  # Year-Week format
    'email_marketing_consent': [100, 120, 150, 180, 200],
    'push_notification_consent': [80, 110, 130, 160, 190]
}

# Create a DataFrame
df = pd.DataFrame(data)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['week'], df['email_marketing_consent'], marker='o', label='Email Marketing Consent')
plt.plot(df['week'], df['push_notification_consent'], marker='s', label='Push Notification Consent')

# Add labels and title
plt.xlabel('Week')
plt.ylabel('Number of Consents')
plt.title('Weekly Progress of Customer Consents')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
